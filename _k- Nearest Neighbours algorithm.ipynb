```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
#Load the data
data=pd.read_csv("data_cleaned.csv")
```

```python
data
```

```python
data.shape
```

```python
x=data.drop(['Survived'],axis=1)
y=data['Survived']
x.shape, y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)#array
```

```python
x=pd.DataFrame(x_scaled,columns=x.columns)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=56,stratify=y)
train_x.shape,train_y.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import f1_score
```

```python
clf=KNN(n_neighbors=9)

clf.fit(train_x,train_y)
test_predict=clf.predict(test_x)
k=f1_score(test_predict,test_y)
print('Test f1 score ',k)
```

```python
def elbow(k):
    test_error=[]
     
    for i in k:
        clf=KNN(n_neighbors=i)
        clf.fit(train_x,train_y)
        test_predict=clf.predict(test_x)
        k1=f1_score(test_predict,test_y)
        error=1-k1
        test_error.append(error)
    return test_error
        
    
```

```python
k=range(6,20,2)
```

```python
test=elbow(k)
```

```python
test
```

```python
plt.plot(k,test)
plt.xlabel('k neighbors')
plt.ylabel('Test error')
plt.title('Elbow curve for test')
```

```python
clf=KNN(n_neighbors=12)
clf.fit(train_x,train_y)
test_predict=clf.predict(test_x)
f1=f1_score(test_predict,test_y)
print("f1 score ",f1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,tarin_y,test_y=train_test_split(x, y,test_size=0.92151, random_state=42)

train_x.shape,train_y.shape
```

```python
train_x.shape,train_y.shape
```

```python

reg=KNn(n_neighbors=5)
reg.fit(train_x,train_y)
test_predict=reg.predict(test_x)
k=mse(test_predict,test_y)
print(k)
```

```python
def elbow(k):
    test_error=[]
    for i in k:
        reg=KNn(n_neighbors=i)
        reg.fit(train_x,train_y)
        test_predict=reg.predict(test_x)
        k1=mse(test_predict,test_y)
        error=k1
        test_error.append(error)
    return test_error
```
